```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import nmc_verification.nmc_vf_base as nvb
import pandas as pd
import numpy as np
import datetime
import copy
```

随机生成包含多个层次，多个时刻，多个时效，多个成员的站点数据，用于测试示例

```python
data = {"id":[54511,54522],"lon":[100,110],"lat":[30,40],"grapes":[0,0],"ec":[0,0],"ncep":[0,0],}
df = pd.DataFrame(data)
sta = nvb.sta_data(df)
nvb.set_stadata_coords(sta,level = 1000,time = datetime.datetime(2019,12,31,8,0),dtime = 0)
sta1 = copy.deepcopy(sta)
nvb.set_stadata_coords(sta1,level = 850)
sta = nvb.combine_join(sta,sta1)
sta1 = copy.deepcopy(sta)
nvb.set_stadata_coords(sta1,time = datetime.datetime(2020,1,1,20,0))
sta = nvb.combine_join(sta,sta1)
sta1 = copy.deepcopy(sta)
nvb.set_stadata_coords(sta1,dtime = 24)
sta_all = nvb.combine_join(sta,sta1)
sta_all.iloc[:,-3:] = (np.random.rand(16,3)*100).astype(np.int16)
print(sta_all)
```

# 按预报天数提取部分时效数据  
**in_dday_list(data,dday_list):**   
从网格和站点数据中提取部分时效的数据，提取时会将 dtime维度的预报时效转换成预报天数，再和列表中的结果进行对比。预报天数 = （预报时效/24小时）向上取整  


**参数说明：**  
 **data**:[站点数据](https://www.showdoc.cc/nmc?page_id=3744334022014027)         
 **dday_list**: 待提取的预报天数序列，0以上的整数        
 **return**: 如果data为站点数据，则返回站点数据，如果data为网格数据，则返回网格数据    


# 按小时余数提取部分时效数据  
**in_dhour_list(data,dhour_list):**   
从网格和站点数据中提取部分时效的数据，提取时会将 dtime维度的预报时效转换成小时余数，再和列表中的结果进行对比。小时余数 = 预报时效整除24小时后的余数  


**参数说明：**  
 **data**:[站点数据](https://www.showdoc.cc/nmc?page_id=3744334022014027)         
 **dhour_list**: 待提取的小时余数序列，0-23的整数        
 **return**: 如果data为站点数据，则返回站点数据，如果data为网格数据，则返回网格数据   

**调用示例**

```python
sta = nvb.in_dtime_list(sta_all,[24])
print(sta)
```

```python
sta = nvb.in_dday_list(sta_all,[1])
print(sta)
```

```python
sta = nvb.in_dhour_list(sta_all,[0])
print(sta)
```

# 通过综合配置字典选取数据  
**by_loc_dict(data,loc_dict):**   
通过包含多个维度选项的字典型参数，来确定所要选取的数据的范围，返回相应的数据样本  
   
**参数说明：**  
 **data**: [站点数据](https://www.showdoc.cc/nmc?page_id=3744334022014027)        
 **loc_dict**: 字典型变量  其形式为：  
 
 
     loc_dict = {'member':["ecmwf","grapes"],  #列表形式
                'level' :[850,700],   #列表形式,
                'time' : [datetime(2020,1,1,8,0)],  #列表形式,
                "time_range":[datetime(2020,1,1,8,0),datetime(2020,1,2,8,0)],  #闭区间,
                "year":[2020,2021], #列表形式，
                "month": [1,2,3]， #列表形式，
                "day": [datetime(2020,1,1,0,0)],  #列表形式，
                "dayofyear":[1,2,365,366] ,  #列表形式，
                "hour":[0,1,23], #列表形式，

                'ob_time' : [datetime(2020,1,1,8,0)],  #列表形式,
                "ob_time_range":[datetime(2020,1,1,8,0),datetime(2020,1,2,8,0)],  #闭区间,
                "ob_year":[2020,2021],  #列表形式，
                "ob_month": [1,2,3]， #列表形式，
                "ob_day": [datetime(2020,1,1,0,0)]， # 列表形式，
                "ob_dayofyear":[1,2,365,366], #列表形式，
                "ob_hour":[0,1,23],  #列表形式，

                'dtime':[24,36]， #列表形式,
                'ddya': [0,1,2]， #列表形式，
                "dhour":[0,1,23], #列表形式，

                'id':[54511]，  #列表形式,
                'lon': [70,140], # 闭区间,
                'lat' :[10,60],  #闭区间,
    }
 
 **return**: 如果data为站点数据，则返回站点数据，如果data为网格数据，则返回网格数据             


**调用示例**


```python
sta = nvb.by_loc_dict(sta_all,{"member":["ec"],"level":[850],"id":[54511],"ob_dayofyear":[1]})
print(sta)
```

```python
sta = nvb.by_loc_dict(sta_all,{"hour":[8],"lon":[105,115]})
print(sta)
```

# 提取非缺省数据  
**not_IV(data)**   
从站点数据中提取不等于默认缺省值的数据样本，若某一行数据中任意一个数据列中包含缺省值，则该行不会被保留    


**参数说明：**  
 **data**: [站点数据](https://www.showdoc.cc/nmc?page_id=3744334022014027)       
 **return**: 如果data为站点数据，则返回站点数据，如果data为网格数据，则返回网格数据   
 

# 提取不等于指定值的数据  
**not_equal_to(data,dele_value)**   
从站点数据中提取不等于指定值的数据样本，若某一行数据中任意一个数据列中包含指定值，则该行不会被保留    


**参数说明：**  
 **data**: [站点数据](https://www.showdoc.cc/nmc?page_id=3744334022014027)   
 **dele_value**: 要删除的样本取值          
 **return**: 如果data为站点数据，则返回站点数据，如果data为网格数据，则返回网格数据   
 

# 提取指定取值范围的数据  
**between_value_range(data,start_value,end_value)**   
从站点数据中提取不等于指定值的数据样本，若某一行数据中任意一个数据列超出范围，则该行不会被保留    


**参数说明：**  
 **data**: [站点数据](https://www.showdoc.cc/nmc?page_id=3744334022014027)       
 **start_value**: 保留样本的最小值   
 **end_value**: 保留样本的最大值，保留的范围是[start_value，end_value] 是闭区间      
 **return**: 如果data为站点数据，则返回站点数据，如果data为网格数据，则返回网格数据   

**调用示例**

```python
sta = nvb.not_IV(sta_all)
print(sta)
```

```python
sta = nvb.not_equal_to(sta_all,0)
print(sta)
```

```python
sta = nvb.between_value_range(sta_all,0,60)
print(sta)
```
