###  列联表  
**contingency_table(ob, fo, grade_list=None, save_path=None)**  
统计观测和预报分别为各种类别时对应的样本数  

**参数说明：**  
 Ob:实况数据，任意维numpy数组   
 Fo:预测数据，任意维numpy数组,Fo.shape 和Ob.shape一致  
 grade_list: 如果该参数为None，观测或预报值出现过的值都作为分类标记，如果该参数不为None，它必须是一个从小到大排列的实数，以其中列出的数值划分出的多个区间作为分类标签。对比预报和观测值不为整数的情况，grade_list 不能设置为None。  
 save_path: 如果该参数不为None,则混淆矩阵将会以excel文件形式保存到指定目录  
 return: 内容为混淆矩阵的二维数组，shape为（类别数 + 1）×（类别数 + 1）  
**调用示例：**  

```python
contingency_table(ob_int,fo_int,save_path = r"H:\test_data\output\nmc_vf_method\multi_catagory\t1.xlsx")
```

在上述示例中，由于没有设置grade_list,所以默认是以观测和预报中所有出现过的值为分类标记，以下是上述运行结果excel表格的截图：

```python
contingency_table(ob_int,fo_int,grade_list,r"H:\test_data\output\nmc_vf_method\multi_catagory\t2.xlsx")
```

###  频率表

 return: 内容为 2×类别数 的二维数组。  
**调用示例：**  

```python
frequency_table(ob_int,fo_int,save_path = r"H:\test_data\output\nmc_vf_method\multi_catagory\t3.xlsx")
```
